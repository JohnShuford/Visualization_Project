```python
import pandas as pd
import json
import os
from pprint import pprint
```

```python
filepath = os.path.join("Data", "RU_category_id.json")
with open(filepath) as jsonfile:
    video_json = json.load(jsonfile)
```

```python
video_json
```

```python
data = video_json['items']
data
```

```python
ru_cat_list = []

for x in data:
    ru_cat_list.append(x['snippet']['title'])
    
ru_cat_list
```

```python
ru_id_list = []

for x in data:
    ru_id_list.append(int(x['id']))
    
ru_id_list
```

```python
ru_cat_df = pd.DataFrame(ru_cat_list, ru_id_list)
ru_cat_df
```

```python
ru_cat_df = ru_cat_df.reset_index()
ru_cat_df
```

```python
ru_cat_df.columns
```

```python
ru_cat_df = ru_cat_df.rename(columns={'index':'category_id', 0:'Category_name'})
ru_cat_df
```

```python
ru_cat_df.to_csv('RU_Categories.csv')
```
